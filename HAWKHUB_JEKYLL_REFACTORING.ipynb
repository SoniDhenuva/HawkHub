````python
# Check for cross-references to classroom-specific posts in navigation and templates

import re

def search_for_references(directory, pattern, file_patterns=['.html', '.yml', '.md']):
    """Search files for references to a pattern"""
    matches = []
    for root, dirs, files in os.walk(directory):
        dirs[:] = [d for d in dirs if not d.startswith('.')]
        for file in files:
            if any(file.endswith(fp) for fp in file_patterns):
                filepath = os.path.join(root, file)
                try:
                    with open(filepath, 'r', encoding='utf-8', errors='ignore') as f:
                        content = f.read()
                        if re.search(pattern, content, re.IGNORECASE):
                            rel_path = os.path.relpath(filepath, directory)
                            matches.append(rel_path)
                except:
                    pass
    return matches

base_dir = '/home/samhi/nighthawk/HawkHub'

# Search for references to classroom features
classroom_refs = {
    'calendar': search_for_references(base_dir, r'calendar'),
    'course|CSA|CSP|CSSE': search_for_references(base_dir, r'course|CSA|CSP|CSSE'),
    'student.toolkit': search_for_references(base_dir, r'student.?toolkit|student_toolkit'),
    'bathroom': search_for_references(base_dir, r'bathroom'),
    'gamify': search_for_references(base_dir, r'gamify'),
}

print("\n" + "=" * 70)
print("POST DEPENDENCIES IN TEMPLATES & DATA FILES")
print("=" * 70)

for feature, files in classroom_refs.items():
    if files:
        print(f"\n{feature.upper()} references ({len(files)} files):")
        for f in files[:5]:
            if '_includes/nav/' in f or '_data/' in f or 'base.html' in f:
                print(f"  ⚠️ {f}")
            else:
                print(f"     {f}")
        if len(files) > 5:
            print(f"  ... and {len(files) - 5} more")
````

````python
# Section 2: Map Liquid Template Includes and Dependencies

import re
import os
from collections import defaultdict

def find_liquid_includes(directory):
    """Recursively find all Liquid {% include %} statements"""
    includes = defaultdict(list)
    
    for root, dirs, files in os.walk(directory):
        # Skip .git and other hidden dirs
        dirs[:] = [d for d in dirs if not d.startswith('.')]
        
        for file in files:
            if file.endswith(('.html', '.md', '.liquid')):
                filepath = os.path.join(root, file)
                try:
                    with open(filepath, 'r', encoding='utf-8', errors='ignore') as f:
                        content = f.read()
                        # Find all {% include ... %} statements
                        matches = re.findall(r'{%\s*include\s+([^%}]+)\s*%}', content)
                        if matches:
                            rel_path = os.path.relpath(filepath, directory)
                            for match in matches:
                                # Extract just the filename
                                include_name = match.split()[0]
                                includes[rel_path].append(include_name)
                except Exception as e:
                    pass
    
    return includes

# Analyze includes
base_dir = '/home/samhi/nighthawk/HawkHub'
includes = find_liquid_includes(os.path.join(base_dir, '_includes'))

print("=" * 70)
print("LIQUID INCLUDE DEPENDENCY MAP")
print("=" * 70)

# Show critical includes
critical_files = [
    '_themes/minima/base.html',
    'themes/minima/base.html',
    '_includes/themes/minima/base.html',
]

# Print base.html includes
base_html_includes = {}
for include_path, include_list in list(includes.items())[:20]:
    if 'base.html' in include_path:
        print(f"\n{include_path}:")
        for inc in include_list:
            print(f"  ├─ includes: {inc}")
            base_html_includes[inc] = include_path
````
